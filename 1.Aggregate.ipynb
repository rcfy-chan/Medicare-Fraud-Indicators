We want to group the rows according to NPI.

For Categorical features, keep unique values, i.e., if HCPCS_Cd is more than one, join the unique entries creating a new entry. 

For Numerical features, Tot_Benes, Tot_Srvcs and Tot_Bene_Day_Srvcs we will sum and for the other numerical features, we are using weighted average with Tot_Bene_Day_Srvcs as weight.

```python
%%time

# Define a custom aggregation function for categorical variables
def concatenate_unique(series):
    unique_values = series.dropna().astype(str).unique()
    return ', '.join(unique_values)

# Define a custom aggregation function for weighted average
def weighted_average(series, weights):
    return np.average(series, weights=weights)

# Group by 'Rndrng_NPI' and aggregate the columns
agg_df = df.groupby('NPI').agg({

    'Last_Org_Name': concatenate_unique,
    'First_Name': concatenate_unique,
    'MI': concatenate_unique,
    'Crdntls': concatenate_unique,
    'Gndr': concatenate_unique,
    'Ent_Cd': concatenate_unique,
    'St1' : concatenate_unique,
    'St2' : concatenate_unique,
    'City': concatenate_unique,
    'State_Abrvtn': concatenate_unique,
    'State_FIPS': concatenate_unique,
    'Zip5': concatenate_unique,
    'RUCA': concatenate_unique,
    'RUCA_Desc': concatenate_unique,
    'Cntry': concatenate_unique,
    'Type': concatenate_unique,
    'Mdcr_Prtcptg_Ind': concatenate_unique,
    'HCPCS_Cd': concatenate_unique,
    'HCPCS_Desc': concatenate_unique,
    'HCPCS_Drug_Ind': concatenate_unique,
    'Place_Of_Srvc': concatenate_unique,    
    'Tot_Benes': 'sum',
    'Tot_Srvcs': 'sum',
    'Tot_Bene_Day_Srvcs': 'sum',
    'Avg_Sbmtd_Chrg': lambda x: weighted_average(x, df.loc[x.index, 'Tot_Bene_Day_Srvcs']),
    'Avg_Mdcr_Alowd_Amt': lambda x: weighted_average(x, df.loc[x.index, 'Tot_Bene_Day_Srvcs']),
    'Avg_Mdcr_Pymt_Amt': lambda x: weighted_average(x, df.loc[x.index, 'Tot_Bene_Day_Srvcs']),
    'Avg_Mdcr_Stdzd_Amt': lambda x: weighted_average(x, df.loc[x.index, 'Tot_Bene_Day_Srvcs']),
}).reset_index()
```
