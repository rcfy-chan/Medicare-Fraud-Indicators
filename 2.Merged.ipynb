### Check if Part B is N and confirm if they are not in df.
### Order and referring data from 2023

```python
order_referring = pd.read_csv('Order_referring.csv')
```

```python
order_referring.head()
```

```python
partb_n = order_referring[order_referring.PARTB == 'N']
partb_n.head()
```

```python
# Create a set of unique NPIs from leie for faster lookup
partb_n_set = set(partb_n['NPI'])

# Create the 'fraud' column using vectorized operations
agg_df['partb_n'] = agg_df['NPI'].isin(partb_n_set).astype(int)
```

```python
merged_df = agg_df.merge(leie[['NPI', 'EXCLTYPE', 'EXCLDATE']], on='NPI', how='left')
```
